```python
# Librer√≠a necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("Gastos y costos 20-23.xlsx", sheet_name='2021', skiprows=3)
data = df
```

```python
data
```

```python
#Identificar valores nulos por columna
data.isnull().sum()
```

```python
# Las columnas que contienen NAs son 'FOLIO' con 147, 'MP' con 654 y 'POLIZA' con 2,372.
```

```python
# Sutituci√≥n de NAs en la columna 'FOLIO' con 147
# Los NAs de la columna 'FOLIO' son un % muy bajo de los registros totales, por ello se sustituyeron los valores faltantes con '000000' para poder identificar que registros no cuentan con un folio en especifico.
data['FOLIO'] = data['FOLIO'].fillna("000000")
data.isnull().sum()
```

```python
# Columna 'MP' con 654
# Se sustituyeron los valores faltantes  de la columna 'MP' con 'No_Especificado'

data['MP'] = data['MP'].fillna("No_Especificado")
data.isnull().sum()
```

```python
# Columna 'POLIZA' con 2372 #
# La columna presentaba  NAs alrededor del 75% de sus registros as√≠ como datos inconsistentes  de caracter texto, num√©rico y fechas mezcladas, por lo que la mejor decisi√≥n es eliminar la columna.  
data = df.drop(columns= ['POLIZA'])
print(data.info())
```

```python
# Se corrobora que los cambios se hayan realizado adecuadamente
data.isnull().sum()
# Confirmando que ya no existen missing values 
```

```python
# Guardar los cambios hechos en un nuevo archivo csv
data.to_csv("Gastos y costos 20-23(2021).xlsx")
```
