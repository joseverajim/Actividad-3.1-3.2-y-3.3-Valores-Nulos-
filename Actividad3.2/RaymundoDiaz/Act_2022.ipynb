```python
#Importar librerías
import pandas as pd
import numpy as np
```

```python
# Ajustar filas y columnas a mostrar
pd.options.display.max_rows = 50
pd.options.display.max_columns = 20
```

```python
# Importar el dataset
df = pd.read_excel('Gastos y costos 20-23.xlsx', sheet_name = '2022', skiprows= 3)
df.info()
```

```python
# Conocer el data set
df
```

```python
# Checar valores nulos
df.isnull().sum()
```

```python
# Primera sustitución de valores nulos para la columna Folio
""" 
Al ser una variable de tipo objeto, se decidió sustituir los 102 valores nulos de la columna, por el valor de folio '5555', con el cual servirá para identificarlos posteriormente.
"""
df_clean = df.copy()
# Checar si hay un folio con el valor de 5555
print(df_clean[df_clean['Folio'] == '5555'])

# Sustitución
df_clean['Folio'] = df_clean['Folio'].fillna('5555')

# Observar la sustitución
print(df_clean.isnull().sum())
```

```python
# Filtrar los registros con MP nulo
filtro = df_clean['MP'].isna()

# Ver si la descripción contiene 'pago' (sin importar mayúsculas/minúsculas)
df_clean.loc[filtro, 'Descripción'].str.contains('pago', case=False, na=False).all()

#df.loc[filtro, ['Descripción', 'MP']] , otra forma de corroborar
```

```python
# Segunda sustitución de valores para la columna MP
"""    
Se decidió sustituir los 552 valores nulos de la columnna MP, por el string 'DescPago', ya que se observa que en la columna Descripción, todos los registros con 'pago', tienen el valor nulo en MP 
"""
df_clean['MP'].value_counts() 

# Susticución con 'DescPago
df_clean['MP'] = df_clean['MP'].fillna('DescPago')


# Observar la sustitución
df_clean.isnull().sum()
```

```python
# Tercera sustitución de valores nulos para la columna TC
print(df_clean['TC'].value_counts())
'''
Se decidió sustituir los 636 valores nulos de la columna TC, por el string '0', debido a que no se encontró una relación entre los registros nulos y cualquier otra columna
'''

df_clean['TC'] = df_clean['TC'].fillna(0)

# Observar la sustitución
df_clean.isnull().sum()
```

```python
# Cuarta sustitución de valores nulos para la columna Otros
'''
Se decidió eliminar la columna Otros y sus 2577 valores nulos, debido a que no aporta información relevante al análisis, ya que no hay ningún solo registro. 
'''

df_final_clean = df_clean.drop(columns = ['Otros'])

# Observar la sustitución
df_final_clean.isnull().sum()
```

```python
# Quinta sustitución de valores nulos para la columna Poliza
'''
Se decidió sustituir los 801 valores nulos de la columna Poliza, utilizando el método forward fill, debido a que las filas con poliza faltante, son anteriores a una fila con poliza, por lo que se asume que será la misma poliza que la anterior.
'''

# Sustitución con método bfill
df_final_clean['Poliza'] = df_final_clean['Poliza'].fillna(method = "ffill")

# Observar la sustitución
df_final_clean.isnull().sum()
```

```python
# Guardar el dataset limpio
df_final_clean.to_csv('Gastos y costos 20-23 Limpio Sheet 2022.csv')
```
